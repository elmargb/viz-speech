# Learning Gaussian Filters

We want to learn feature extractors for the input in terms of Gaussian filters with varying width height and rotations.

Let the Gaussian function be,
$$ g(\mathbf{x};\mathbf{\mu},\mathbf{C}) = \exp\left(-\left(\mathbf{x} - \mathbf{\mu} \right)^{\top} \mathbf{C} \left(\mathbf{x} - \mathbf{\mu} \right)\right) $$

Then we define a weight matrix $\mathbf{W}$ such that,

$$ \underbrace{\mathbf{W}}_{(N,H)} = \mathbf{G} \mathbf{M}$$

$$W_{(i,j)} = \alpha_j g(\left[\text{row}(i),\text{col}(i)\right];\mathbf{\mu}^{(j)},\mathbf{C}^{(j)})$$

where $\text{row}(i)$ and $\text{col}(i)$ give the row and column of input $i$.

In this way, we are describing a layer where all of the neurons have weights with Gaussian looking intensities if they are mapped into a two-dimensional space using the row and column function. Then instead of having a 1024 by 1024 matrix, we are left with the following free parameters: $\mu_j, \sigma_j, \alpha_j$


Result needs to be a (25,4) matrix
