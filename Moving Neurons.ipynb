```python
import numpy as np
import scipy.linalg
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.cm as cm


matplotlib.rcParams['savefig.dpi'] = 4 * matplotlib.rcParams['savefig.dpi']
size = 5
precision = np.eye(2)
mean = np.zeros(2) + (size-1)/2.
print mean


points = np.asarray(np.dstack(np.meshgrid(np.arange(size), np.arange(size))).reshape(-1,2),dtype=np.int32)

g = np.exp(-np.sum(np.dot(points-mean,precision)**2,axis=1))
print g.reshape((size,size))
full = np.zeros((32,32))
full[:size,:size] = g.reshape((size,size))
circ =  scipy.linalg.circulant(np.arange(32)).T
plt.imshow(full[circ[1]],interpolation='nearest')
```
