```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(10,8))
left_right = np.array([-2.5,2.5])
ax = plt.axes(xlim=(left_right[0],left_right[1]), ylim=(-2.5, 2.5))
line, = ax.plot([], [], lw=2)
points = np.asarray(4 * np.random.rand(1000,2) - 2)

ceil = np.ceil(points) + 1
label = (ceil[:,0] % 2 == ceil[:,1] % 2)
class1 = points[ label]
class2 = points[~label]

plt.plot(class1[:,0],class1[:,1], 'ro')
plt.plot(class2[:,0],class2[:,1], 'bo')
```
