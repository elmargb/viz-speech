```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from IPython.html.widgets import interact,IntSliderWidget,RadioButtonsWidget
from IPython.display import clear_output, display, HTML
import grid_learn
import matplotlib
from collections import Counter
matplotlib.rcParams['savefig.dpi'] = 2 * matplotlib.rcParams['savefig.dpi']
```

```python
instances = 1000
grid_size = 3
line_count = 4
train,predict,hidden_p,params = grid_learn.build_model(line_count)
```

```python
points, label = grid_learn.create_data(grid_size,instances)
fig,ax,ax_hist,left_right,labels,lbl_colors,rects,ctr_pts,im,Z = grid_learn.plot(points,label,line_count)
lines = [ax.plot([], [], lw=2)[0] for _ in xrange(line_count)]
acc_text = ax.text(0.02, 1.01, '', transform=ax.transAxes)
acc_text.set_text('')
```

```python
def data_gen():
    acc,coeffs,bias,hidden = train(points,label)

    while acc < 0.999:
        #print acc
        yield acc,coeffs,bias,hidden
        acc,coeffs,bias,hidden = train(points,label)
    for _ in xrange(100):
        #print acc
        yield acc,coeffs,bias,hidden
        acc,coeffs,bias,hidden = train(points,label)

it = data_gen()
```

```python
data = None
for _ in xrange(2500):
    try:
        data = it.next()
    except StopIteration:
        break
```

```python
def redraw(hidden_neuron=0):
    fig,ax,ax_hist,left_right,labels,lbl_colors,rects,ctr_pts,im,Z = grid_learn.plot(points,label,line_count)
    
    lines = [ax.plot([], [], lw=2)[0] for _ in xrange(line_count)]
    acc_text = ax.text(0.02, 1.01, '', transform=ax.transAxes)
    acc_text.set_text('')
    acc,coeffs,bias,hidden = data
    acc_text.set_text("accuracy=%0.2f"%acc)
    
    # draw lines
    M = - coeffs[0]/coeffs[1]
    C = - bias / coeffs[1]
    for i in xrange(len(lines)):
        lines[i].set_data(left_right, M[i]*left_right + C[i])
        
    # draw heatmap
    if hidden_neuron == "output":
        ctr_z = predict(np.dstack(ctr_pts).reshape(-1, 2))
    else:
        ctr_z = hidden_p(np.dstack(ctr_pts).reshape(-1, 2),hidden_neuron)
    im.set_data(ctr_z.reshape(ctr_pts[0].shape))
    activations = Counter(''.join(str(e) for e in row) for row in hidden)
    total = sum(activations.values())
    print total
    #sort labels
    labels.sort(key=lambda x:-activations.get(x,0))
    for i in ax_hist.get_xticklabels(): i.set_color(lbl_colors[i.get_text()])
    for lbl,rct in zip(labels,rects):
        rct.set_height(activations.get(lbl,0)/float(total))
```

```python
interact(
    redraw,
    hidden_neuron=RadioButtonsWidget(values=range(line_count) + ["output"], value=0)
)
```
